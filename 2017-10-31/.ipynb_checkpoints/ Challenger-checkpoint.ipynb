```python
# バイアス項の追加
```

```python
def logistic(x, beta, alpha=0):
    return 1.0/(1.0 + np.exp(beta * x + alpha))

x = np.linspace(-4, 4, 100)

plt.plot(x, logistic(x, 1), label=r"$\beta = 1$", ls="--", lw=1)
plt.plot(x, logistic(x, 3), label=r"$\beta = ３$", ls="--", lw=1)
plt.plot(x, logistic(x, -5), label=r"$\beta = -5$", ls="--", lw=1)

plt.plot(x, logistic(x, 1, 1), label=r"$\beta = 1, \alpha = 1$", color="#348ABD")
plt.plot(x, logistic(x, 3, -2), label=r"$\beta = 3, \alpha = -2$", color="#348ABD")
plt.plot(x, logistic(x, -5, 7), label=r"$\beta = -5, \alpha = 7$", color="#348ABD")

plt.title("Logistic function for "r"different $\beta$ and $\alpha$ values")
plt.xlabel("$x$")
plt.ylabel("Logistic function at $x$") # ロジスティック関数値
plt.legend(loc="lower left")
```

```python
# 正規分布
import scipy.stats as stats
nor = stats.norm
colors = ["#348ABD", "#A60628", "#7A68A6"]

x = np.linspace(-8, 7, 150)
mu = (-2, 0, 3)
tau = (.7, 1, 2.8)
parameters = zip(mu, tau, colors)

for _mu, _tau, _color in parameters:
    plt.plot(x, nor.pdf(x, _mu, scale=1. / _tau),
             label=r"$\mu = %d, \tau = %.1f$" % (_mu, _tau),
             color=_color)
    plt.fill_between(x, nor.pdf(x, _mu, scale=1. / _tau),
                    color=_color, alpha=.33)
    
plt.legend(loc="upper right")
plt.xlabel("$x$")
plt.ylabel("Density function at $x$") # 密度関数値
plt.title("Probability distribution of three different Nomal random variables")
```

```python
# チャレンジャー号のモデリング
```

```python
import pymc as pm

temperature = challenger_data[:, 0]
D = challenger_data[:, 1] #破損発生かどうか

beta = pm.Normal("beta", 0, 0.001, value=0)
alpha = pm.Normal("alpha", 0, 0.001, value=0)

@pm.deterministic
def p(t=temperature, alpha=alpha, beta=beta):
    return 1.0 / (1. + np.exp(beta * t + alpha))
```

```python
p.value
```

```python
# p の確率と観測データをベルヌーイ分布で結びつける
observed = pm.Bernoulli("bernoulli_obs", p, value=D, observed=True)

model = pm.Model([observed, beta, alpha])

# これ以降は3章で
map_ = pm.MAP(model)
map_.fit()
mcmc = pm.MCMC(model)
mcmc.sample(120000, 100000, 2)
```

```python
%pylab inline
figsize(12.5, 6)

alpha_samples = mcmc.trace('alpha')[: , None] #1次元にする
beta_samples = mcmc.trace('beta')[: , None]

plt.subplot(211)
plt.title("Posterior distributions of "r"the model parameters $\alpha, \beta$")
plt.hist(beta_samples, histtype='stepfilled',
        bins=35, alpha=0.85, color="#7A68A6", normed=True,
        label=r"posterior of $\beta$")  # beta の事後分布
plt.legend()

plt.subplot(212)
plt.hist(alpha_samples, histtype='stepfilled',
        bins=35, alpha=0.85, color="#A60628", normed=True,
        label=r"posterior of $\alpha$") # alphaの事後分布
plt.xlabel("Value of parameter") # パラメータ値
plt.ylabel("Density") # 密度
plt.legend()
```

```python
# 期待確率をみる
```

```python
figsize(12.5, 4)

t=np.linspace(temperature.min() - 5, temperature.max() + 5 , 50)[:, None]
p_t = logistic(t.T, beta_samples, alpha_samples)

mean_prob_t = p_t.mean(axis=0)

plt.plot(t, mean_prob_t, lw=3,
        label="average posterior \n probability of defect") #破損の平均事後確率
plt.plot(t, p_t[0, :], ls="--", label="realization from posterior") #事後分布からのサンプル
plt.plot(t, p_t[-2, :], ls="--", label="realization from posterior") #事後分布からのサンプル

plt.scatter(temperature, D, color="k", s=50, alpha=0.5)

plt.title("Posterior expected value of the probability of defect, including two realizations")
plt.legend(loc="lower left")
plt.ylim(-0.1, 1.1)
plt.xlim(t.min(), t.max())
plt.ylabel("Probability") #確率
plt.xlabel("Temperature") #外気温
```

```python
from scipy.stats.mstats import mquantiles

# 信頼区間の上下2.5%
qs = mquantiles(p_t, [0.025, 0.975], axis=0)
plt.fill_between(t[:, 0], *qs, alpha=0.7, color="#7A68A6")

plt.plot(t[:, 0], qs[0],label="95% CI", color="#7A68A6", alpha=0.7) # 95% 信頼区間
plt.plot(t, mean_prob_t, lw=1, ls="--", color="k", 
         label="average posterior probability of defect") # 破損の平均確率
plt.xlim(t.min(), t.max())
plt.ylim(-0.02, 1.02)
plt.legend(loc="lower left")

plt.scatter(temperature, D, color="k", s=50, alpha=0.5)
plt.xlabel("Temperature, $t$") # 外気温
plt.ylabel("Probability estimate") # 確率の推定値
plt.title("Posterior probability of estimates, given temperature $t$")
```

```python
# 2.3 モデルは適切？？
simulated = pm.Bernoulli("bernoulli_sim", p)
N = 10000

mcmc = pm.MCMC([simulated, alpha, beta, observed])
mcmc.sample(N)

simulations = mcmc.trace("bernoulli_sim")[:].astype(int)

# 生成データarrayのサイズ
print("Shape of simylations array : ", simulations.shape)
plt.suptitle("Simulated datasets using posterior parameters")

figsize(12.5, 20)
for i in range(4):
    ax = plt.subplot(4, 1, i+1)
    plt.scatter(temperature, simulations[1000*i, :], color="k", s=50, alpha=0.6)
```

```python
# セパレーションプロット

posterior_probability = simulations.mean(axis=0)

# 観測， 破損発生の有無を生成したデータ， 破損発生の事後確率， 破損発生
print("Obs. | Array of Simulated Defects | Posterior Probability of Defect | Realized Defect")

print(D)

for i in range(len(D)):
    print("%-4s | %s | %-12.2f | %d" %
         (str(i).zfill(2),
         str(simulations[:10, i])[:-1] + "...]".ljust(6),
         posterior_probability[i], D[i]))
```

```python
ix = np.argsort(posterior_probability)

# 破損発生の事後確率， 破損発生
print("Posterior Probability of Defect | Realized Defect")
for i in range(len(D)):
    print("%-31.2f | %d" % (posterior_probability[ix[i]], D[ix[i]]))
```
